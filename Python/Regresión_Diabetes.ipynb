```python
import pandas as pd
import kagglehub
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Descarga del dataset
path = kagglehub.dataset_download("mathchi/diabetes-data-set")

print("Path to dataset files:", path)

# Construir el path completo al archivo CSV
file_path = f"{path}/diabetes.csv"

# Importar el dataset como un dataframe de Pandas
df = pd.read_csv(file_path)
df.info()
df
```

```python
# Detección y tratamiento de valores faltantes o erróneos
#1. Detección

# df.isnull() devuelve un DataFrame con True/False según haya nulos
# df.isnul().sum() devuelve el conteo de nulos por columna
print(" Valores faltantes:")
print(df.isnull().sum())


#2. Visualización
#Una vez identificados los valores nulos
#Procedemos a visualizarlos


sns.heatmap(df.isnull(), cbar=False) # Graficamos un mapa de calor (heatmap) para visualizar nulos
plt.show() # Permite mostrar el gráfico
print("No se presentan valores faltantes")
```

```python
#Análisis descriptivo de variables numéricas
import matplotlib.pyplot as plt
import seaborn as sns

numericas = ['Pregnancies','Glucose', 'BloodPressure','SkinThickness','Insulin', 'BMI','DiabetesPedigreeFunction', 'Age']

for col in numericas:
    print(f"\n Análisis descriptivo numérico: {col}")
    print(df[numericas].describe().round(2))
    print(f"Asimetría: {df[col].skew():.2f}")
    print(f"Curtosis: {df[col].kurtosis():.2f}")

    # Histograma + KDE
    plt.figure(figsize=(6, 3))
    sns.histplot(df[col], kde=True, color='lightblue')
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()

    # Boxplot
    plt.figure(figsize=(6, 2))
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f'Boxplot de {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

```python
# Se consideran solo las personas con diabetes para analizar la relación entre variables
df_diabetes = df[df["Outcome"] == 1]
print(df.isnull().sum())
df_diabetes
```

```python
# Correlación entre variables numéricas
corr_matrix = df_diabetes.corr() #por defecto usa pearson

print(corr_matrix)

# Visualizar con heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f") #annot permite incluir los valores en el gráfico
plt.title('Matriz de correlación entre variables que influyen en Diabetes')
plt.show()
```

```python
#independencia de errores
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Variable dependiente y explicativa
X = df['Glucose'] #Varable independiente o explicativa o predictora
y = df['Pregnancies'] #Variable dependiente o respuesta

# Agregar constante (intercepto) para el modelo
X_const = sm.add_constant(X) #escalar que intercepta en eje Y
modelo = sm.OLS(y, X_const).fit()

# Mostrar resumen del modelo
print(modelo.summary())

# Predicción y residuos
df['Prediccion'] = modelo.predict(X_const) #cálculo de los valores predichos y crea una columna llamada predicción
df['Residuos'] = y - df['Prediccion'] # Calcula y guarda los residuos en una columna llamada residuos
```

```python
#Comportamiento lineal
plt.figure(figsize=(6,5))
sns.regplot(x=df['Glucose'], y=df['Pregnancies'], line_kws={'color': 'red'})
plt.title("Relación lineal entre Glucosa y Embarazos")
plt.xlabel("Glucosa")
plt.ylabel("Embarazos")
plt.show()
```
